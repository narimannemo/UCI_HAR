```python
import pandas as pd
import tensorflow as tf
import numpy as np
from tensorflow import keras
from tensorflow.keras.layers import Dense, Dropout, Conv1D, Flatten
from tensorflow.keras.models import Sequential
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical
```

```python
train_data=pd.read_csv('../../Datasets/UCI_HAR/train.csv')
test_data=pd.read_csv('../../Datasets/UCI_HAR/test.csv')
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
X=train_data.iloc[:,:-2]
y=train_data.iloc[:,-1]

X_test=test_data.iloc[:,:-2]
y_test=test_data.iloc[:,-1]
```

```python
X_test.shape
```

```python
X = X.values.reshape(7352,561)
X_test=X_test.values.reshape(2947,561)
```

```python
labelencoder_y_1 = LabelEncoder()
labelencoder_y_test_1 = LabelEncoder()

y = labelencoder_y_1.fit_transform(y)
y_test = labelencoder_y_test_1.fit_transform(y_test)


y_binary = to_categorical(y)
y_binary_test = to_categorical(y_test)

X = X.reshape(X.shape[0], 561, 1)
X_test = X_test.reshape(X_test.shape[0], 561, 1)
```

```python
model = Sequential()
model.add(Conv1D(64, kernel_size=3, activation='relu', input_shape=(561,1)))
model.add(Conv1D(32, kernel_size=3, activation='relu'))
model.add(Flatten())
model.add(Dense(6, activation='softmax'))
model.compile(optimizer='sgd',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(X,y_binary,epochs=23)
```

```python
y_prediction = model.predict_classes(X_test)
print ("\n\naccuracy", np.sum(y_prediction == y_test) / float(len(y_test)))
```
